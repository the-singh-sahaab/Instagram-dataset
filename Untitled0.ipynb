```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor


data =  pd.read_csv("/content/Instagramdata.csv", encoding = 'Latin1')
print(data)
```

checking for the null values and droping them 

```python
data.isnull().sum()
```

```python
data.info()
```

visualizating


```python
plt.figure(figsize = (7,7))
plt.style.use("fivethirtyeight")
plt.title("Distribution of impresssions from home")
sns.distplot(data['From Home'])
plt.show()
```

```python
plt.figure(figsize = (5,7))
plt.style.use("fivethirtyeight")
plt.title("Distribution of impresssions from explore")
sns.distplot(data['From Explore'])
plt.show()
```

```python
home = data["From Home"].sum()
hashtags = data["From Hashtags"].sum()
explore = data["From Explore"].sum()
other = data["From Other"].sum()

labels = ['From Home','From Hashtags','From Explore','Other']
values = [home, hashtags, explore, other]

fig = px.pie(data, values=values, names=labels, title='Impressions on Instagram Posts From Various Sources', hole=0.5)
fig.show()
```

```python
figure = px.scatter(data_frame = data, x = 'Impressions', y = 'Likes', size = 'Likes', trendline = 'ols', title = 'Relationship Between Likes and Impression')
figure.show()
```

```python
correlation = data.corr()
print(correlation['Impressions'].sort_values(ascending = False))
```

```python
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']])
y = np.array(data['Impressions'])
xtrain, x_test, ytrain, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

trying with passive aggressive rregressor

```python
model = PassiveAggressiveRegressor()
model.fit(xtrain,ytrain)
model.score(x_test, y_test)
```

trying with ridge mothed

```python
from sklearn import linear_model

model1 = linear_model.Ridge(alpha = 0.5)
model1.fit(xtrain, ytrain)
model1.score(x_test, y_test)
```

trying with liearn regression method

```python
from sklearn import linear_model

model2 = linear_model.LinearRegression()
model2.fit(xtrain, ytrain)
model2.score(x_test, y_test)
```

```python
from sklearn import linear_model

model3 = linear_model.RidgeClassifier()
model3.fit(xtrain, ytrain)
model3.score(x_test, y_test)
```

```python
from sklearn import linear_model
model4 = linear_model.Lasso(alpha = 0.1)
model4.fit(xtrain,ytrain)
model4.score(x_test, y_test)
```

```python
from sklearn import linear_model
model5 = linear_model.LassoLars(alpha = 0.1)
model5.fit(xtrain,ytrain)
model5.score(x_test, y_test)
```

```python
from sklearn import linear_model
model4 = linear_model.BayesianRidge()
model4.fit(xtrain,ytrain)
model4.score(x_test, y_test)
```
